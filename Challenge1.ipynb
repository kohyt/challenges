```python
# Write a program that will accept a date of birth and determine whether one 
# is considered an adult (i.e. 21 years old).
dob = input("Enter date of birth(DD/MM/YYYY): ")
def check_dob(dob):
    if len(dob) == 10:
        if dob[3:5] == "02":
            if int(dob[0:2]) <= 28 or int(dob[0:2]) >= 1:
                return True
            return False
        elif dob[3:5] == "04" or dob[3:5] == "06" or dob[3:5] == "09" or dob[3:5] == "011":
            if int(dob[0:2]) >= 1 or int(dob[0:2]) <= 30:
                return True
            return False
        else:
            if int(dob[0:2]) >= 1 or int(dob[0:2]) <= 31:
                return True
            return False
    else:
        return False
if check_dob(dob):
    age = 2017 - int(dob[-4:])
    print(age)
    if age >= 21:
        print("Adult")
    else:
        print("Not an adult.")
else:
    print("INVALID")
```

```python
# Make a game of rock, paper, scissors against the computer.
# Algorithm
# Tell user to enter either rock,paper or scissors
# Get the response
# Generate a random number from 1 to 3 (1=rock,2=paper,3=scissors)
# Compare user selection and computer selection
# Display who wins.
import random
moves = {1:"rock", 2:"paper", 3:"scissors"}
user = input("Enter move (rock, paper, scissors): ")
comp = moves[random.randint(1,3)]
user = user.lower()
if user == comp:
    print("DRAW")
elif (user == "rock" and comp == "scissors") or (user == "paper" and comp == "rock") or (user == "scissors" and comp == "paper"):
    print("You win!")
else:
    print("You lose!")
```

```python
# For the rock, paper, scissors game above, ensure that the user enters a valid 
# entry. Add a loop structure to play several times and keep a running score.
def check(user):
    if user == "rock" or user == "paper" or user == "scissors":
        return True
    else:
        return False
import random
moves = {1:"rock", 2:"paper", 3:"scissors"}
times = int(input("Enter number of times: "))
time = 0
while time < times:
    user = input("Enter move (rock, paper, scissors): ")
    comp = moves[random.randint(1,3)]
    user = user.lower()
    if check(user):
        time += 1
        if user == comp:
            print("DRAW")
        elif (user == "rock" and comp == "scissors") or (user == "paper" and comp == "rock") or (user == "scissors" and comp == "paper"):
            print("You win!")
        else:
            print("You lose!")
    else:
        print("invalid input. Try again.")
```

```python
# Write a program that will give the answer to logic gate questions e.g.
# Enter logic gate : OR
# Enter first input : 1
# Enter second input :0
# Result = 1
# It should work for the logic gates OR, XOR, NAND and NOR gates.
gate = input("Enter Logic gate : ")
gate = gate.upper()
first = int(input("Enter first input : "))
second = int(input("Enter second input : "))
if gate == "OR":
    if (first == 1 and second == 0) or (first == 0 and second == 1):
        print("Result = 1")
    else:
        print("Result = {}".format(first))
elif gate == "XOR":
    if (first == 1 and second == 0) or (first == 0 and second == 1):
        print("Result = 1")
    else:
        print("Result = 0")
elif gate == "NAND":
    if (first == 1 and second == 0) or (first == 0 and second == 1) or (first == 0 and second == 0):
        print("Result = 1")
    else:
        print("Result = 0")
elif gate == "NOR":
    if first == 0 and second == 0:
        print("Result = 1")
    else:
        print("Result = 0")
else:
    print("Only either OR, XOR, NAND and NOR gates!")
```

```python
# Write a program that will display all the factors of an input number that
# are bigger than 1.
# (e.g. the factors of the number 12 are 6,4,3 and 2 because they divide
# into 12 exactly).
# Hint: To find out whether a number X is a factor of Y use:
# If Y mod X == 0 (there is nothing remaining when Y is divided by X)
def factors(num):
    factors = []
    factor = 2
    while num > 0 and factor < num:
        if num % factor == 0:
            factors.append(factor)
            factor += 1
        else:
            factor += 1
    print(factors)

num = int(input("Enter number: "))
factors(num)
```
